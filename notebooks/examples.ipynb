```python
Plot(['sin(x)', 'sin(2*x)', 'sin(3*x)'], ['x', 0, 2 * pi], legends=True);
plt.axis('equal');
```

```python
Plot3D('sin(x + y**2)', ['x', -3, 3], ['y', -2, 2]);
plt.axis('equal');
```

```python
ParametricPlot([['2*cos(u)', '2*sin(u)'], ['2*cos(u)', 'sin(u)'], ['cos(u)', '2*sin(u)'],\
                ['cos(u)', 'sin(u)']], ['u', 0, 2*pi], legends=True);
plt.axis('equal');
```

```python
PolarPlot(['1', '1 + 1/10 * sin(10*t)'], ['t', 0, 2*pi]);
plt.axis('equal');
```
